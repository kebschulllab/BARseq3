```python
# import packages
import os
import pickle
from starmap.config import Config
from starmap.pipeline import Pipeline
```

# Config Setting
there are two choices for setting up a config
- start from fresh (type in all the params by yourself for your first run)
- initialize with existing saved config from your last run

```python
# choice 1. start from fresh
config = Config(filepath_homedir = "",
                filepath_rawdata = "",
                filepath_codebook = "",
                filepath_dapi = "",
                filepath_nissl = "",
                filepath_dapi_all_z = "",
                round_index = [1, 2, 3], 
                fov_align = 1,
                round_align = 1,
                fov_minmax = [1, 3, 5, 6, 10],
                cellpose_model = 'TN3',
                cellpose_channel = [2, 3],
                SpGene = None, # 'ChAlign', 'RoAlign', 'CbCorrection', 'Normalization', 'bardensrReg', 'GeneCalling', 'Finished'
                SpMask = None, # 'PosReg', 'CellSeg', 'ExpMask', 'Gene2Mask', 'RmvOverlap', 'Finished'
                
                ifov = 1,
                igenes = ["gene1", "gene2", "gene3", "gene4", "nogene1", "nogene2"]
                )
```

```python
# choice 2. load saved config
filename = 'filepath of config.pickle stored in validation folder'
with open(filename, 'rb') as f:
    config_dict = pickle.load(f)
config = Config(**config_dict)
```

```python
# update several setting if needed
config.update_parameters()
```

# Pipline Initialization
here pipeline are auto loaded and prepared for run

```python
pipeline = Pipeline(config)
```

# Run Pipeline
actually run the pipeline through

```python
pipeline.run()
if config.CpMask['RmvOverlap'] and config.remove_overlap:
    if config.grid_stitch:
        os.system('python remove_overlap_mp.py ' + config.filepath_homedir + ' position_reg ' + str(config.dim))
    else:
        os.system('python remove_overlap_mp.py ' + config.filepath_homedir + ' position_corr ' + str(config.dim))
```
